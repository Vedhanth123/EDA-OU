```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
a= pd.read_csv("ToyotaCorolla.csv")
a.head()
```

```python
a.tail()
```

```python
a.info()
```

```python
a.describe()
```

```python
a['Fuel_Type'].value_counts().plot(kind='bar')
plt.xlabel('fuel')
plt.ylabel('count')
plt.title('visualisation of fuel vs count')
plt.show()
```

```python
sns.scatterplot(x='Mfg_Year', y='Age_08_04', data=a)
```

```python
sns.boxplot(x='HP', y='KM', data=a)
```

```python
correlation_matrix = a.corr()
sns.heatmap(correlation_matrix, annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
a['Sport_Model'].value_counts().plot(kind='kde')
plt.xlabel('Sport_model')
plt.ylabel('count')
plt.show()
```

```python
price = a['Price']
age = a['HP']
km = a['Mfg_Year']
fig, ax = plt.subplots()
ax.plot(price, label='Price')
ax.plot(age, label='HP')
ax.plot(km, label='Mfg_Year')
```

```python
sns.violinplot(x=a['Fuel_Type'], y= a['Price'])
```

```python
columns = ['Price', 'KM', 'HP']

# Create a pairplot using seaborn
sns.pairplot(a[columns])

# Display the plot
plt.show()
```
