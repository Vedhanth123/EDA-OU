```python
df['Survived'].replace({0:'Not Survived', 1:'Survived'}, inplace=True)
df.tail()
```

```python
x = pd.crosstab(df['Pclass'],df['Survived'], margins=True)
print(x)
```

```python
x.iloc[:-1,:-1].plot(kind='bar',stacked=False, grid=False, title='Survival␣Analysis Per Class')
plt.show()
```

```python
sns.catplot(y=df['Age'], x=df['Sex'])
```

```python
sns.set(color_codes=True)
sns.distplot(df['Age'].dropna())
```

```python
sns.set(style='ticks',color_codes=True)
sns.pairplot(df, vars=['Pclass','Fare', 'Age'], hue='Survived')
```

```python
sns.jointplot(data=df, x="Age", y="Fare", kind="scatter")
```

```python
sns.scatterplot(x=df['Age'], y=df['Fare'], alpha=0.5, color='mediumvioletred')
plt.xlabel('Age')
plt.ylabel('Fare')
```

```python
df.boxplot(column='Age', by='Survived',showmeans=True)
```
